```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/customer_retention.csv")  # Replace with actual file name

# Display first few rows
print(df.head())

###  Step 1: Identify Root Causes of Churn ###
# Assume 'Segment' represents churn behavior (e.g., 1 = Retained, 0 = Churned)
df['Churn'] = df['Segment'].apply(lambda x: 1 if x == 0 else 0)

# Analyzing correlations
correlation = df[['Age', 'Income', 'SpendingScore', 'Tenure', 'InternetUsage', 'Churn']].corr()
print("\nCorrelation with Churn:")
print(correlation['Churn'].sort_values(ascending=False))

### Step 2: Data Visualization (Finding Patterns) ###
plt.figure(figsize=(12, 6))

# üîπ Churn vs Income
plt.subplot(1, 2, 1)
sns.boxplot(x='Churn', y='Income', data=df, palette='coolwarm')
plt.title("Income Distribution by Churn")
plt.xlabel("Churn (1 = Left, 0 = Stayed)")
plt.ylabel("Income")

# üîπ Churn vs Spending Score
plt.subplot(1, 2, 2)
sns.boxplot(x='Churn', y='SpendingScore', data=df, palette='coolwarm')
plt.title("Spending Score Distribution by Churn")
plt.xlabel("Churn (1 = Left, 0 = Stayed)")
plt.ylabel("Spending Score")

plt.tight_layout()
plt.show()

###  Step 3: Retention Strategies ###
# Insights based on data patterns
strategies = """
1Ô∏è‚É£ **High-Spending Customers with Low Tenure**:
   - Introduce loyalty programs & early incentives.
   - Offer exclusive deals to build engagement.

2Ô∏è‚É£ **Low-Income, High-Churn Customers**:
   - Provide flexible payment plans & discounts.
   - Improve customer experience & communication.

3Ô∏è‚É£ **Low Internet Usage Customers**:
   - Improve digital engagement strategies.
   - Offer educational support on digital tools.

4Ô∏è‚É£ **Older Customers with High Churn**:
   - Personalized outreach to address concerns.
   - Special retention programs for senior users.
"""

print("\ n **Retention Strategies:**")
print(strategies)

###  Step 4: Business Recommendations (Report Format) ###
report = f"""
# üìä **Customer Churn Analysis Report**
## **Key Findings**
- Customers with **lower income** tend to **churn more**.
- Short tenure customers with **high spending** are at risk of leaving.
- **Low internet users** show higher churn rates.

## **Recommendations**
1Ô∏è‚É£ Focus on high-risk customer segments.
2Ô∏è‚É£ Implement **personalized retention offers**.
3Ô∏è‚É£ Use **data-driven marketing** to engage customers.
4Ô∏è‚É£ Enhance **customer support & communication**.

### **Next Steps**
üîπ Implement targeted **loyalty programs**.  
üîπ Offer **discounts & personalized pricing**.  
üîπ Improve **digital engagement & service quality**.
"""

# Save Report as a text file
with open("Customer_Churn_Report.txt", "w", encoding="utf-8") as f:

    f.write(report)

print("\ n  **Report Generated:** 'Customer_Churn_Report.txt'")
```
