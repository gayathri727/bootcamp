```python
# Step 1: Install required libraries (if not installed)
!pip install wordcloud matplotlib nltk pandas

# Step 2: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re  # Alternative method for tokenization

# Step 3: Download stopwords and tokenizer
nltk.download('stopwords')
nltk.download('punkt')

# Step 4: Load the dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/ipl_commentary_v2.csv")

# Step 5: Combine all the commentary text
text = " ".join(comment for comment in df["Commentary"])

# Step 6: Remove stopwords
stop_words = set(stopwords.words('english'))

try:
    # Try using word_tokenize
    words = word_tokenize(text)
except LookupError:
    # Alternative method if word_tokenize() fails
    print("word_tokenize() failed, using regex instead.")
    words = re.findall(r'\b\w+\b', text)  # Tokenize using regex

# Step 7: Filter out stopwords
filtered_text = " ".join(word for word in words if word.lower() not in stop_words)

# Step 8: Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="coolwarm").generate(filtered_text)

# Step 9: Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Hide axes
plt.show()
```
