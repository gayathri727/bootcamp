```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/life_expectancy.csv")

# Set plot style
sns.set_style("whitegrid")

# Create the plot
plt.figure(figsize=(8, 6))
sns.lineplot(x=df["Life Expectancy"], y=df["Year"], marker="o", color="b", linestyle="-")

# Labels and title
plt.xlabel("Average Life Expectancy (Years)")
plt.ylabel("Year")
plt.title("Life Expectancy Over Time")

# Show plot
plt.show()
```
