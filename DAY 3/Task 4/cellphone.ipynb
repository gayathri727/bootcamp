```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/cell_phones_data.csv")

# Plotting
plt.figure(figsize=(10,6))
plt.plot(df['Cell Phones per Person'], df['Year'], marker='o', color='orange', label='Cell Phones per Person')
plt.title('Number of Cell Phones per Person vs Year')
plt.xlabel('Number of Cell Phones per Person')
plt.ylabel('Year')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
