```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/happiness_data.csv")

# Plotting
plt.figure(figsize=(10,6))
plt.plot(df['Happiness Score (WHR)'], df['Year'], marker='o', color='purple', label='Happiness Score (WHR)')
plt.title('Happiness Score (WHR) vs Year')
plt.xlabel('Happiness Score (WHR)')
plt.ylabel('Year')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
