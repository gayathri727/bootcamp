```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/bubble_chart_data.csv")

# Filter data for a specific year (Example: 2020)
year = 2020
df_year = df[df["Year"] == year]

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=df_year["Income"], 
    y=df_year["Life Expectancy"], 
    size=df_year["Population"], 
    hue=df_year["Country"], 
    sizes=(50, 1000), 
    alpha=0.6, 
    edgecolor="w"
)

plt.xscale("log")  # Log scale for better visualization
plt.xlabel("Income (GDP per capita)")
plt.ylabel("Life Expectancy (Years)")
plt.title(f"Life Expectancy vs Income ({year})")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
