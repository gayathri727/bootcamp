```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("c:/Users/shruthi/Downloads/income_data_30.csv")

# Set plot style
sns.set_style("whitegrid")

# Create the plot
plt.figure(figsize=(9, 6))
sns.lineplot(x=df["Average Income"], y=df["Year"], marker="o", color="g", linestyle="-")

# Labels and title
plt.xlabel("Average Income ($)")
plt.ylabel("Year")
plt.title("Average Income Over Time (Extended Data)")

# Show plot
plt.show()
```
